```python
data_comp = data.groupby(data['company']).count().reset_index()
data_comp
```

```python
data_comp.plot('company', 'position', kind = 'bar')
```

```python
top_50= data_comp.nlargest(50, 'position')
```

```python
top_50= data_comp.nlargest(50, 'position').drop(columns=['description', 'url'])
```

```python
company = top_50['company']
no_jd = top_50['position']


plt.barh(company, no_jd)
plt.title('Jobs by Company')
plt.ylabel('Company')
plt.xlabel('No.')
plt.show()
```
