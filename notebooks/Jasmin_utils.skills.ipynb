```python
from fydjob.NLPFrame import NLPFrame
from fydjob.utils import count_skills
import pandas as pd
```

```python
df = NLPFrame().df
df
```

```python
count = count_skills(df['skills'])
count.keys()
```

```python
df_skills = pd.DataFrame()

for key in count.keys():
    print(key)
    df_key=pd.DataFrame.from_dict(count[key])
    df_key['Category']=key
    df_skills = df_skills.append(df_key)
    
df_skills=df_skills.rename(columns = {0:'Skill',1:'Count'})
df_skills
```
