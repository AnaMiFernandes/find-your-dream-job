```python
def get_similarities(text, text_vector):
    similarities = []
    for i, t in enumerate(text_vector):
        sim = len(set(text) & set(t)) / (len(set(text)) + len(set(t))) * 2
        similarities.append((i, sim))
    similarities = sorted(similarities, key=lambda x: x[1], reverse=True)
    return similarities 
```

```python
job_sims = {}

for i, job in enumerate(df[:50].job_text_tokenized):
    sims = [x for x in get_similarities(job, df.job_text_tokenized)
           if x[1] > .95 and x[0] != i]
    job_sims[i] = sims
```
